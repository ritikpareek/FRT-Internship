```python
# importing the libraries
import selenium
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException,ElementNotVisibleException,ElementNotInteractableException
import pandas as pd
import numpy as np
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

# Question 1 Solution

```python
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
rank=[]
name=[]
artist=[]
date=[]
views=[]
```

```python
try:
    r=driver.find_elements_by_xpath("//td[@align='center'][1]")[0:30]
    for i in r:
        R=i.text
        rank.append(R)
except NoSuchElementException:
    rank.append('-')
print(rank)
```

```python
try:
    n=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div/table[1]/tbody/tr/td[2]")[0:30]
    for i in n:
        N=i.text
        name.append(N)
except NoSuchElementException:
    name.append('-')
print(name)
```

```python
try:
    a=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div/table[1]/tbody/tr/td[3]")[0:30]
    for i in a:
        A=i.text
        artist.append(A)
except NoSuchElementException:
    artist.append('-')
try:
    d=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div/table[1]/tbody/tr/td[5]")[0:30]
    for i in d:
        D=i.text
        date.append(D)
except NoSuchElementException:
    date.append('-')
    
try:
    v=driver.find_elements_by_xpath("//*[@id='mw-content-text']/div/table[1]/tbody/tr/td[4]")[0:30]
    for i in v:
        V=i.text
        views.append(V)
except NoSuchElementException:
    views.append('-')
    
```

```python
print(len(rank),len(name),len(views),len(date),len(artist))
```

```python
df1=pd.DataFrame({'rank':rank,'name':name,'artist':artist,'views':views,'date':date})
df1
```

# Question 10 Solution

```python
url='https://archive.ics.uci.edu/ml/index.php'
driver.get(url)
```

```python
driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]").click()
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
```

```python
try:
    n=driver.find_elements_by_xpath("//p[@class='normal']/b/a")
    for i in n:
        N=i.text
        Dataset_name.append(N)
except NoSuchElementException:
    Dataset_name.append('-')
print(Dataset_name)
```

```python
try:
    dt=driver.find_elements_by_xpath("/html/body/table/tbody/tr/td/table/tbody/tr/td[2]/p")[2:]
    for i in dt:
        Dt=i.text
        Data_type.append(Dt)
except NoSuchElementException:
    Data_type.append('-')
print(Data_type)
```

```python
try:
    t=driver.find_elements_by_xpath("/html/body/table/tbody/tr/td/table/tbody/tr/td[3]/p")[1:]
    for i in t:
        T=i.text
        Task.append(T)
except NoSuchElementException:
    Task.append('-')
try:
    at=driver.find_elements_by_xpath("/html/body/table/tbody/tr/td/table/tbody/tr/td[4]/p")[1:]
    for i in at:
        At=i.text
        Attribute_type.append(At)
except NoSuchElementException:
    Attribute_type.append('-')
    
try:
    ins=driver.find_elements_by_xpath("/html/body/table/tbody/tr/td/table/tbody/tr/td[5]/p")[1:]
    for i in ins:
        Ins=i.text
        No_of_instances.append(Ins)
except NoSuchElementException:
    No_of_instances.append('-')
    
try:
    atr=driver.find_elements_by_xpath("/html/body/table/tbody/tr/td/table/tbody/tr/td[6]/p")[1:]
    for i in atr:
        Atr=i.text
        No_of_attribute.append(Atr)
except NoSuchElementException:
    No_of_attribute.append('-')


try:
    y=driver.find_elements_by_xpath("/html/body/table/tbody/tr/td/table/tbody/tr/td[7]/p")[1:]
    for i in y:
        Y=i.text
        Year.append(Y)
except NoSuchElementException:
    Year.append('-')
```

```python
print(len(Dataset_name))
print(len(Data_type))
print(len(Task))
print(len(Attribute_type))
print(len(No_of_instances))
print(len(No_of_attribute))
print(len(Year))
```

```python
df10=pd.DataFrame({'dname':Dataset_name,'dtype':Data_type,'task':Task,'attr':Attribute_type,'inst':No_of_instances,'noattr':No_of_attribute,'year':Year})
df10.head()
```

# Question 5 Solution

```python
url='https://github.com/'
driver.get(url)
```

```python
driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a").click()
```

```python
title=[]
des=[]
count=[]
lang=[]
```

```python
try:
    t=driver.find_elements_by_xpath("//span[@class='text-normal']")[1:]
    for i in t:
        T=i.text.replace('/',' ')
        title.append(T)
except NoSuchElementException:
    title.append('-')
print(title)
```

```python
try:
    d=driver.find_elements_by_xpath("//p[@class='col-9 text-gray my-1 pr-4']")[1:]
    for i in d:
        D=i.text.replace('/',' ')
        des.append(D)
except NoSuchElementException:
    des.append('-')
try:
    c=driver.find_elements_by_xpath("/html/body/div/main/div/div/div/article/div/a[2]")[1:]
    for i in c:
        C=i.text.replace('/',' ')
        count.append(C)
except NoSuchElementException:
    count.append('-')
try:
    l=driver.find_elements_by_xpath("//span[@itemprop='programmingLanguage']")[1:]
    for i in l:
        L=i.text.replace('/',' ')
        lang.append(L)
except NoSuchElementException:
    lang.append('-')
```

```python
print(len(title),len(des),len(count),len(lang))
```

```python
df5=pd.DataFrame({'titile':title[:20],'des':des[:20],'count':count[:20],'language':lang[:20]})
df5
```

# Question 9 Solution

```python
url='https://www.imdb.com/list/ls095964455/'
driver.get(url)
```

```python
name=[]
year=[]
genre=[]
time=[]
rating=[]
votes=[]
```

```python
try:
    n=driver.find_elements_by_xpath("//*[@id='main']/div/div/div/div/div/h3/a")
    for i in n:
        N=i.text
        name.append(N)
except NoSuchElementException:
    name.append('-')
try:
    y=driver.find_elements_by_xpath("//span[@class='lister-item-year")
    for i in y:
        Y=i.text
        Year.append(Y)
except NoSuchElementException:
    Year.append('-')
try:
    d=driver.find_elements_by_xpath("//span[@class='certificate']")
    for i in d:
        D=i.text
        genre.append(D)
except NoSuchElementException:
    genre.append('-')
try:
    r=driver.find_elements_by_xpath("//*[@id='main']/div/div/div/div/div[2]/div[1]/div[1]/span[2]")
    for i in r:
        R=i.text
        rating.append(R)
except NoSuchElementException:
    rating.append('-')
try:
    v=driver.find_elements_by_xpath("//span[@name='nv']")
    for i in v:
        V=i.text
        votes.append(V)
except NoSuchElementException:
    votes.append('-')
```

```python
df9=pd.DataFrame({'name':name[:100],'year':Year[:100],'genre':genre[:100],'rating':rating[:100],'votes':votes[:100]})
df9
```
