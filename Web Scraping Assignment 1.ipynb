# Book Name

```python
page=requests.get('https://bookpage.com/reviews?book_genre=fiction&page=1')
```

```python
page.status_code
```

```python
page.text
```

```python
soup=BeautifulSoup(page.text,"html.parser")
```

```python
soup.head
```

```python
book_name=[]
author_name=[]
genre=[]
review=[]
```

```python
a=soup.findAll('h4',class_='italic')[0:5]
```

```python
a
```

```python
for i in a:
    book_name.append(i.get_text())
```

```python
book_name
```

```python
book_name = list(map(str.strip,book_name))
```

```python
book_name
```

```python
b=soup.findAll('p',class_='sans bold')[0:5]
b
```

```python
for i in b:
    author_name.append(i.get_text())
```

```python
author_name
```

```python
author_name = list(map(str.strip,author_name))
```

```python
author_name
```

```python
c=soup.findAll('p',class_='genre-links hidden-phone')[0:5]
c
```

```python
for i in c:
    genre.append(i.get_text())
```

```python
genre
```

```python
d=soup.findAll('div',class_='row-fluid')[0:5]
d
```

```python
for i in d:
    review.append(i.get_text())
```

```python
review
```

```python
review= list(map(str.strip,review))
```

```python
review
```

```python
df3=pd.DataFrame({'b_name':book_name,'a_name':author_name,'genre':genre,'review':review})
```

```python
df3.to_csv('books_data.csv')
```
